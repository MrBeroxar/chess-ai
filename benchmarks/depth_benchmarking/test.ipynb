```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bestmove_benchmarkings.csv")
df = df[df["nodes"] != 0]
```

```python
df.head(20)
```

```python
def plot():
    grouped = df.groupby("nodes")
    plt.figure(figsize=(8, 6))
    for name, group in grouped:
        timestamp = group["date"] 
        average_time = group["average_time"]
        label = f"Nodes: {name}"
        plt.plot(timestamp, average_time, label=label)  
    plt.xlabel("date")  
    plt.ylabel("Average Time in ms")
    plt.title("Average Time (Grouped by Nodes/Depth)")  
    plt.legend()
    plt.grid(True)
    plt.show()

plot()
```
